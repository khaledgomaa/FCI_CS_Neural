```python
import pandas as pd
import numpy as np

data=pd.read_csv('/home/gomaa/Desktop/titanic_data.csv')

data.head()

```

```python
columns_target=['Survived']


columns_train=['Age','Pclass','Sex','Fare']

X=data[columns_train]
Y=data[columns_target]
#check if there is any null values to clean our data 
X['Sex'].isnull().sum()

X['Pclass'].isnull().sum()

X['Fare'].isnull().sum()

X['Age'].isnull().sum() 

#I got null vlaues with Age column so we need to modify this null values
#we can use median to don't loss our data
X['Age']=X['Age'].fillna(X['Age'].median())
X['Age'].isnull().sum()
#now we filled all the null values of Age column with median value 
#we need to make our dataset numerical which can be calculated
#we start with sex values to replace male with 0 and female with 1
d={'male':0 ,'female':1}
X['Sex']=X['Sex'].apply(lambda x:d[x])
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.33,random_state=42)

from sklearn import svm
clf=svm.LinearSVC()
clf.fit(X_train,Y_train)
print (clf)

```

```python
print (clf.predict(X_test[0:1]))

print (clf.predict(X_test[0:10]))

print (clf.score(X_test,Y_test))
```
